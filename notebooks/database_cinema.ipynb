```python
import pandas as pd

csv_file_path_people = "../data/af-am-film-data-1.2/people.csv"
csv_file_path_films = "../data/af-am-film-data-1.2/films.csv"
csv_file_path_companies = "../data/af-am-film-data-1.2/companies.csv"

df_people = pd.read_csv(csv_file_path_people)


df_films = pd.read_csv(csv_file_path_films)
df_companies = pd.read_csv(csv_file_path_companies) 

df_companies.head()

```

```python
df_films_companies = pd.merge(right=df_films, 
                                   left=df_companies, how="inner", 
                                   right_on="Production Company", left_on="Name")

#df_films_companies.info()
df_films_companies.describe()
```

```python
import geopandas as gpd
import descartes
import matplotlib.pyplot as plt
import pandas as pd
from shapely.geometry import Point, Polygon

%matplotlib inline

street_map =gpd.read_file('../data/af-am-film-data-1.2/world_map/World_Countries.shp')
fig,ax = plt.subplots(figsize =(15,15))
street_map.plot(ax=ax)

crs ={'init':'epsg:4326'}
geometry = [Point(xy) for xy in zip(df_films_companies["Longitude"],df_films_companies["Latitude"])]
geometry[:3]

geo_df = gpd.GeoDataFrame(df_films_companies,crs=crs,geometry = geometry)
geo_df.head()


```

```python
fig,ax = plt.subplots(figsize =(15,15))
street_map.plot(ax=ax,alpha =0.4,color='grey')
geo_df[geo_df['WmvPresent']==0].plot(ax=ax,markersize =20,color='blue',marker='o',label='Neg')
geo_df[geo_df['WmvPresent']==1].plot(ax=ax,markersize =20,color='blue',marker='^',label='Pos')
plt.legend(prop={'size':15})
```

```python
import pandas as pd

csv_file_path_people = "../data/af-am-film-data-1.2/people.csv"
csv_file_path_films = "../data/af-am-film-data-1.2/films.csv"
csv_file_path_companies = "../data/af-am-film-data-1.2/companies.csv"

df_people = pd.read_csv(csv_file_path_people)
df_people.set_index('Full name', drop=True, inplace=True)

df_films = pd.read_csv(csv_file_path_films) 
#df_films.set_index('Title', drop=True, inplace=True)

df_companies = pd.read_csv(csv_file_path_companies) 
df_companies.set_index('Name', drop=True, inplace=True)
```

```python
df_companies["Company_name"] = df_companies.index

df_companies.head()
df_companies.groupby("White or AfAm Owned")["Company_name"].count()
```

```python
df_films['Production Company']= [str(company).split(',') for company in df_films["Production Company"].tolist()]
df_films.head()
```

```python



#create dictionnary 
films_company = {"films_title":list(),"production_company":list()}
for title, companies in df_films[["Title", "Production Company"]].values:
    for company in companies:
        films_company["films_title"].append(title)
        films_company["production_company"].append(company)
        
        
df_films_company = pd.DataFrame(films_company)
```

```python
df_films["Title"] = df_films.index

df_films.head()
df_films_genre = df_films.groupby("Genre")["Title"].count()
df_films_genre = pd.DataFrame(df_films_genre)
df_films_genre = df_films_genre.rename(columns={"Title": "number_films"})
df_films_genre = df_films_genre[df_films_genre["number_films"] > 2]
#df.groupby(['n_mentions'])[['text']].count()
```
